```python
mapping = {
    'CP': 'CONFIRMED',
    'KP': 'CONFIRMED',
    'PC': 'CANDIDATE',
    'FP': 'REJECTED',
    'FA': 'REJECTED'
}

tess_imputed['target'] = tess_copy['tfopwg_disp'].map(mapping)
```

```python
print(tess_imputed['target'].value_counts())
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
x = tess_imputed.drop(['target', 'tfopwg_disp'], axis=1)
y = tess_imputed['target']
```

```python
X = x.select_dtypes(include=['number'])
print(f"Columnas numricas: {X.shape[1]}")
print(f"Filas: {X.shape[0]}")
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("\nOriginal Classes:", label_encoder.classes_)
print("Conversion:", dict(zip(label_encoder.classes_, range(len(label_encoder.classes_)))))
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
print(f"\nX_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
```

```python
param_grid = {
    'n_estimators': [200, 300, 500],
    'max_depth': [5, 7, 9, 11],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.8, 0.9, 1.0],
    'min_child_weight': [1, 3, 5],
    'gamma': [0, 0.1, 0.2]
}
```

```python
grid = GridSearchCV(
    xgb.XGBClassifier(objective='multi:softprob', num_class=3, random_state=42),
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
```

```python
grid.fit(X_train, y_train)
```

```python
print("Best parameters:", grid.best_params_)
print("Best score:", grid.best_score_)
```

```python
model = xgb.XGBClassifier(
    objective='multi:softprob',
    num_class=3,
    learning_rate=0.05,
    n_estimators=1000,
    max_depth=7,
    eval_metric='mlogloss',
    random_state=42
)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy", accuracy)
```

```python
print(classification_report(y_test, y_pred))
```
